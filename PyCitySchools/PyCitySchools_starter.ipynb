```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
num_schools = school_data['school_name'].count()

# Calculate the total number of students
num_students = student_data['student_name'].count()

# Calculate the total budget
total_budget = school_data['budget'].sum()

# Calculate the average math score
avg_math = student_data['math_score'].mean()

# Calculate the average reading score
avg_reading = student_data['reading_score'].mean()

# Calculate the percentage of students with a passing math score (70 or greater)
math_pass = student_data[student_data['math_score'] >= 70].count()['student_name']
percent_math_pass = (math_pass / num_students) * 100

# Calculate the percentage of students with a passing reading score (70 or greater)
reading_pass = student_data[student_data['reading_score'] >= 70].count()['student_name']
percent_reading_pass = (reading_pass / num_students) * 100

# Calculate the percentage of students who passed math and reading (% Overall Passing)
num_pass_both = student_data[(student_data['math_score'] >= 70) & (student_data['reading_score'] >= 70)].count()['student_name']
percent_pass_both = (num_pass_both / num_students) * 100

# Create a dataframe to hold the above results
district_summary = pd.DataFrame(
    {"Total Schools": [num_schools],
    "Total Students": num_students,
    "Total Budget": total_budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_reading,
    "% Passing Math": percent_math_pass,
    "% Passing Reading": percent_reading_pass,
    "% Overall Passing": percent_pass_both})
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# group by school
schools_group = school_data_complete.groupby("school_name")
```

```python
# find the passing percentage for each school
per_math = []
per_reading = []
per_overall = []
names = []

for name,group in schools_group:
    names.append(name)
    current_school = schools_group.get_group(name)
    students = current_school['student_name'].count()
    
    math = current_school[current_school['math_score'] >= 70].count()['student_name']
    percent_math_passing = (math / students) * 100
    per_math.append(percent_math_passing)
    
    reading = current_school[current_school['reading_score'] >= 70].count()['student_name']
    percent_reading_passing = (reading / students) * 100
    per_reading.append(percent_reading_passing)

    overall = current_school[(current_school['math_score'] >= 70) & (current_school['reading_score'] >= 70)].count()['student_name']
    percent_pass_overall = (overall / students) * 100
    per_overall.append(percent_pass_overall)
```

```python
passing_percentage = pd.DataFrame({
    'school_name': names,
    'per_math': per_math,
    'per_reading': per_reading,
    'per_overall': per_overall
})
passing_percentage_groups = passing_percentage.groupby("school_name")
```

```python
# create series for each column
schools_type = schools_group.type.nth(0)
total_students = schools_group.student_name.count()
total_school_budget = schools_group.budget.nth(0)
per_student_budget = total_school_budget / total_students
average_math_score = schools_group.math_score.mean()
average_reading_score = schools_group.reading_score.mean()
per_math_pass = passing_percentage_groups.per_math.sum()
per_reading_pass = passing_percentage_groups.per_reading.sum()
per_overall_pass = passing_percentage_groups.per_overall.sum()
```

```python
# Create a dataframe to hold the above results
school_summary = pd.DataFrame({'School Type': schools_type,
                               'Total Students': total_students,
                               'Total School Budget': total_school_budget,
                               'Per Student Budget': per_student_budget,
                              'Average Math Score': average_math_score,
                              'Average Reading Score': average_reading_score,
                              '% Passing Math': per_math_pass,
                              '% Passing Reading': per_reading_pass,
                              '% Overall Passing': per_overall_pass})
school_summary
```
